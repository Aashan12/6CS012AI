# Exercise 2: Image Processing with Pillow, NumPy and Matplotlib

## Task 1: Load and Display a Grayscale Image

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image using Pillow
img = Image.open("cameraman.png").convert("L")
img_array = np.array(img)

print("Image shape:", img_array.shape)
print("Data type:", img_array.dtype)
print("Min pixel value:", img_array.min())
print("Max pixel value:", img_array.max())

# Display the grayscale image
plt.figure(figsize=(6, 6))
plt.imshow(img_array, cmap="gray")
plt.title("Task 1: Original Grayscale Image")
plt.axis("off")
plt.show()
```

## Task 2: Extract and Display the Middle Section (150 pixels)

```python
# Get image dimensions
height, width = img_array.shape

# Calculate center crop coordinates (150x150 from center)
center_y, center_x = height // 2, width // 2
half_crop = 150 // 2

# Extract the middle 150x150 section using NumPy slicing
cropped = img_array[center_y - half_crop : center_y + half_crop,
                    center_x - half_crop : center_x + half_crop]

print("Cropped image shape:", cropped.shape)

# Display the cropped image
plt.figure(figsize=(5, 5))
plt.imshow(cropped, cmap="gray")
plt.title("Task 2: Middle 150px Section")
plt.axis("off")
plt.show()
```

## Task 3: Apply a Simple Threshold (Binary Image)

```python
# Apply threshold: pixels below 100 -> 0, pixels above 100 -> 255
threshold = 100
binary_img = np.where(img_array < threshold, 0, 255).astype(np.uint8)

print("Unique values in binary image:", np.unique(binary_img))

# Display the binary image
plt.figure(figsize=(6, 6))
plt.imshow(binary_img, cmap="gray")
plt.title("Task 3: Binary Image (Threshold = 100)")
plt.axis("off")
plt.show()
```

## Task 4: Rotate the Image 90 Degrees Clockwise

```python
# Rotate 90 degrees clockwise using NumPy
# np.rot90 rotates counter-clockwise, so k=3 gives clockwise 90
rotated_img = np.rot90(img_array, k=3)

print("Original shape:", img_array.shape)
print("Rotated shape:", rotated_img.shape)

# Display the rotated image
plt.figure(figsize=(6, 6))
plt.imshow(rotated_img, cmap="gray")
plt.title("Task 4: Rotated 90° Clockwise")
plt.axis("off")
plt.show()
```

## Task 5: Convert Grayscale Image to RGB

```python
# Convert grayscale to RGB by stacking the same channel 3 times
rgb_img = np.stack([img_array, img_array, img_array], axis=2)

print("Grayscale shape:", img_array.shape)
print("RGB shape:", rgb_img.shape)

# Display the RGB image
plt.figure(figsize=(6, 6))
plt.imshow(rgb_img)
plt.title("Task 5: Grayscale Converted to RGB")
plt.axis("off")
plt.show()
```
